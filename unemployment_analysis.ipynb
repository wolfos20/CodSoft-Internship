### PLOTTING THE UNEMPLOYMENT GRAPH BASED ON REGION

```python

plt.figure(figsize=(17,5))
fig, ax = plt.subplots()
diag = plt.bar(df1['Region'], df1[' Estimated Unemployment Rate (%)'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.xlabel('Region')
1
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by Region')
plt.show()
```

```python
#ENCODING THE DATASET1
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df1['Region'] = label_encoder.fit_transform(df1['Region'])

df1_numeric = df1.select_dtypes(include=[float, int])
corr_matrix = df1_numeric.corr()
```

```python
# plotting correlation heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

### PLOTTING THE UNEMPLOYMENT GRAPH BASED ON REGION

```python
plt.figure(figsize=(17, 9))
fig, ax = plt.subplots()
diag = plt.bar(df2['Region'], df2[' Estimated Unemployment Rate (%)'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.xlabel('Region')
1
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by Region')
plt.show()
```

```python
#ENCODING THE DATASET2
label_encoder = LabelEncoder()
df2['Region'] = label_encoder.fit_transform(df2['Region'])

df2_numeric = df2.select_dtypes(include=[float, int])


corr_matrix = df2_numeric.corr()
```

```python
# plotting correlation heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
#PLOTTING HISTPLOT FOR ESTIMATED EMPLOYED
df1.columns=['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region']
plt.title('Indian Unemployment')
sns.histplot(data=df1,x='Estimated Employed',hue='Region',palette="inferno",bins=20)
plt.show()
```
